#git使い方
##gitリポジトリの初期化
プロジェクトをgitで管理  
→プロジェクトのリポジトリに移動  
→＜git init＞コマンドを実行  
⇒gitリポジトリが初期化され、gitによるバージョン管理ができる  
##ファイルを管理対象に追加
→ファイルを作成  
→＜git add＞コマンドを実行  
→＜git commit＞コマンドを実行  
⇒作成したファイルをgitの管理対象に加えられる  
（例）＄ echo Hello > README.md  
　　　＄ git add README.md  
　　　＄ git commit -m "first commit"   
　　　　 ※ -mの後の文字列はそのコミットについてのメモ
##編集履歴の記録・確認
####記録
→管理対象ファイルを変更  
→＜git add＞コマンドを実行  
→＜git commit＞コマンドを実行
⇒管理対象ファイルの変更が記録される  
※＜git status＞コマンドで差分のあるファイルを確認できる  
※＜git diff＞コマンドで変更内容を確認できる  
　　※＜git add＞済みのファイルは《--cached》オプションが必要  
####確認
→＜git log＞コマンドを実行  
⇒これまで記録したコミットログが確認できる  
※黄色字で表示されているのが、コミットハッシュ(コミットを一意に識別するためのもの)
##チェックアウト
→＜git checkout コミットハッシュ＞コマンドを実行  
⇒履歴をさかのぼって特定のバージョン時点の状態にする  
※元々の状態に戻すには＜git checkout main＞コマンドを実行  
##作業のやり直し
####間違えて＜git add＞してしまったファイルを取り消す
→（例）＜git reset HEAD ファイル名＞  
→＜git reset＞  
　　※一括で取り消す
####間違えて行ったコミットの取り消し
→（例）＜git reset HEAD＞
##ブランチとマージ
####ブランチ
時系列を分岐して記録していくためのもの
####マージ
ブランチの切り替え
